## 3. Machine Learning
####  Q1: Can you describe the difference between supervised and unsupervised learning?

####  A1: Supervised learning involves training a model on a labeled dataset, where the correct outputs are known. Unsupervised learning, in contrast, involves training a model on a dataset without labeled responses, typically used to identify patterns or groupings in the data.
####  Q2: How do you evaluate a machine learning model’s performance?

####  A2: The evaluation depends on the type of model (classification, regression, etc.). Common metrics include accuracy, precision, recall, and F1-score for classification models, and mean squared error (MSE) or mean absolute error (MAE) for regression models.
####  Q3: What is cross-validation in machine learning, and why is it used?

####  A3: Cross-validation is a technique used to assess the generalizability of a model by dividing the dataset into multiple subsets, training the model on several subset combinations, and validating it on the remaining parts. It helps prevent overfitting and provides a more robust measure of model performance.
####  Q4: Explain the concept of overfitting. How can it be avoided?

####  A4: Overfitting occurs when a model learns the detail and noise in the training data to an extent that it negatively impacts the performance of the model on new data. It can be avoided by simplifying the model, using more data, or applying techniques like regularization.
####  Q5: What are gradient descent and its variants?

####  A5: Gradient descent is an optimization algorithm used to minimize the cost function in many machine learning algorithms. Variants like stochastic gradient descent (SGD) and mini-batch gradient descent provide improvements in speed and stability by updating the model's parameters using only a subset of the data for each iteration.
####  Q6: Describe a situation where you would use a random forest over a decision tree.

####  A6: Random forest is typically used over a single decision tree when you need higher accuracy and better model generalization. A random forest reduces overfitting by averaging multiple decision trees, each trained on different parts of the same training set.
####  Q7: What is feature scaling, and why is it important?

####  A7: Feature scaling involves standardizing the range of independent variables or features of data. It is important because many machine learning algorithms perform better or converge faster when features are on a relatively similar scale and close to normally distributed.
####  Q8: How do you handle imbalanced datasets in classification problems?

####  A8: Techniques include resampling the dataset (either by oversampling the minority class or undersampling the majority class), using anomaly detection techniques, and applying different cost functions to penalize wrong predictions on the minority class more than those on the majority class.
####  Q9: What are precision and recall?

####  A9: Precision is the ratio of correctly predicted positive observations to the total predicted positives. It is a measure of accuracy provided that a class label has been predicted. Recall, or sensitivity, is the ratio of correctly predicted positive observations to all observations in the actual class.
####  Q10: Can you explain what ensemble learning is?

####  A10: Ensemble learning is a technique that combines several base models in order to produce one optimal predictive model. Techniques like boosting, bagging, and stacking are used to increase the accuracy of models by reducing variance, bias, or improving predictions.

## 4. Asynchronous Programming
####  Q1: What is asynchronous programming and why is it used in Python?

####  A1: Asynchronous programming allows for the handling of multiple tasks seemingly at once, by executing them in an overlapping manner. In Python, this is often done using the asyncio library. It is used for I/O-bound and high-level structured network code.
####  Q2: Can you explain the difference between synchronous and asynchronous execution?

####  A2: Synchronous execution refers to executing tasks sequentially, blocking on each operation until it completes. Asynchronous execution involves tasks running in overlap with each other, not waiting for each to complete before moving on to the next task, thereby improving efficiency.
####  Q3: What are coroutines in Python, and how do they work with async/await?

####  A3: Coroutines are functions in Python that can suspend their execution before reaching return, and they can indirectly pass control to other coroutines for some time. They are defined with async def and are used with await to pause the coroutine, waiting for an operation to complete without blocking the event loop.
####  Q4: How does the asyncio event loop work?

####  A4: The asyncio event loop runs tasks, managing the execution of several tasks via scheduling, using an internal queue for tasks that are ready to run, while tasks waiting for external operations may be paused until their operations are completed.
####  Q5: Explain the asyncio.Future object.

####  A5: A Future is an object that represents an eventual result of an asynchronous operation. Unlike futures in other languages, asyncio.Future is not suitable for waiting for results in concurrent code; instead, await is used.
####  Q6: What is a task in asyncio, and how is it different from a future?

####  A6: A task is a subclass of Future designed to wrap a coroutine. While a Future can be created from any function by third-party libraries, a Task specifically manages the execution state and result of a coroutine, making coroutine-based concurrency easier to implement.
####  Q7: How do you handle exceptions in asynchronous code in Python?

####  A7: Exceptions in asynchronous Python code can be handled using try-except blocks around the await expressions. Additionally, asyncio provides mechanisms to set exception handlers on the event loop.
####  Q8: What are the common uses of aiosqlite and aiofiles?

####  A8: aiosqlite allows for asynchronous interaction with SQLite databases, enabling non-blocking database calls. aiofiles is used for asynchronous file handling, allowing file read/write operations to be non-blocking.
####  Q9: Can you provide an example of how to use anyio to create a TCP server?

####  A9: anyio provides an abstraction layer for compatibility across multiple asynchronous event loop implementations. A TCP server can be created using anyio by defining an asynchronous function that handles client connections and then starting a server that listens for incoming connections.
####  Q10: How do you test asynchronous Python code?

####  A10: Testing asynchronous code typically requires a test framework that supports asyncio, such as pytest with its pytest-asyncio plugin. Tests need to be written as asynchronous functions using async def and executed using await within the test functions.

#### Q2: Merge Two Sorted Lists

##### Problem: Merge two sorted lists into a single sorted list without using any built-in sort methods.

```python
def merge_sorted_lists(list1, list2):
    result = []
    i = j = 0
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i += 1
        else:
            result.append(list2[j])
            j += 1
    result.extend(list1[i:])
    result.extend(list2[j:])
    return result
```

#### Q3: Decorator for Execution Time

##### Problem: Write a Python decorator that measures and prints the execution time of a function.

```python
import time 
import numpy as np

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        func(*args, **kwargs)
        end = time.time()
        print(func.__name__, f"Took: {end-start:.2f} seconds")

    return wrapper

@timer
def example_function(np_array):

    return np.where(np_array > 0, "Active", "Inactive")


np_array = np.random.randn(1000000)
example_function(np_array)
```

#### Q2: Asynchronous API Calls

##### Problem: Use aiohttp to make asynchronous HTTP GET requests to fetch data from multiple URLs and print the responses as they arrive.

```python
import aiohttp
import asyncio

async def fetch(url, session):
    async with session.get(url) as response:
        return await response.text()

async def fetch_all(urls):
    async with aiohttp.ClientSession() as session:
        tasks = [fetch(url, session) for url in urls]
        results = await asyncio.gather(*tasks)
        return results

# urls = ['http://example.com', 'http://example.org']
# asyncio.run(fetch_all(urls))
```

#### Q3: Scheduling Asynchronous Tasks

##### Problem: Write a Python script that uses asyncio to schedule and run two asynchronous tasks that print "Task 1 Completed" and "Task 2 Completed" after random delays.

```python
import asyncio
import random

async def task1():
    await asyncio.sleep(random.randint(1, 3))
    print("Task 1 Completed")

async def task2():
    await asyncio.sleep(random.randint(1, 3))
    print("Task 2 Completed")

async def main():
    await asyncio.gather(task1(), task2())

await main()
```

### Multiprocessing and Multithreading
#### Problem Description
##### Given a list of numbers from 1 to NN (where NN is large, say 1,000,000), compute the sum of the square roots of all numbers in the list. Implement this: Using multithreading / Using multiprocessing.

```python
import threading
import math
from queue import Queue

def worker(num_queue, result_queue):
    total = 0
    while not num_queue.empty():
        num = num_queue.get()
        total += math.sqrt(num)
        num_queue.task_done()
    result_queue.put(total)

def sum_sqrt_multithreading(numbers, num_threads):
    num_queue = Queue()
    result_queue = Queue()
    threads = []

    for number in numbers:
        num_queue.put(number)
    
    for i in range(num_threads):
        thread = threading.Thread(target=worker, args=(num_queue, result_queue))
        thread.start()
        threads.append(thread)
    
    for thread in threads:
        thread.join()
    
    total_sum = 0
    while not result_queue.empty():
        total_sum += result_queue.get()

    return total_sum

# Usage
N = 1000
numbers = range(1, N + 1)
result = sum_sqrt_multithreading(numbers, 10)
print("Multithreading result:", result)
```

```python
import multiprocessing
import math

def worker(nums, out_q):
    """ The worker function, invoked in a process. Sum up the square roots. """
    out_sum = sum(math.sqrt(num) for num in nums)
    out_q.put(out_sum)

def sum_sqrt_multiprocessing(numbers, num_processes):
    """ Divide the work into number of processes """
    process_list = []
    out_q = multiprocessing.Queue()
    chunksize = int(len(numbers) / num_processes)
    for i in range(num_processes):
        p = multiprocessing.Process(
            target=worker,
            args=(numbers[chunksize * i:chunksize * (i + 1)], out_q)
        )
        process_list.append(p)
        p.start()
    
    for p in process_list:
        p.join()

    total_sum = 0
    while not out_q.empty():
        total_sum += out_q.get()
    
    return total_sum

# Usage
N = 1000
numbers = list(range(1, N + 1))  # list needed for slicing in multiprocessing
result = sum_sqrt_multiprocessing(numbers, 10)
print("Multiprocessing result:", result)
```

```python
import pandas as pd
import numpy as np

# Settings
num_objects = 5  # Number of distinct objects
num_records_per_object = 10  # Number of records per object
total_records = num_objects * num_records_per_object
date_range = pd.date_range(start='2024-01-01', periods=num_records_per_object, freq='H')

# Generate data
np.random.seed(42)  # For reproducibility
object_ids = np.repeat(np.arange(1, num_objects + 1), num_records_per_object)
times = np.tile(date_range, num_objects)
x_coords = np.random.normal(loc=0, scale=1, size=total_records).cumsum(axis=0)
y_coords = np.random.normal(loc=0, scale=1, size=total_records).cumsum(axis=0)
z_coords = np.random.normal(loc=0, scale=1, size=total_records).cumsum(axis=0)

# Create DataFrame
data = pd.DataFrame({
    'ObjectID': object_ids,
    'Time': times,
    'X': x_coords,
    'Y': y_coords,
    'Z': z_coords
})


data.set_index('ObjectID', inplace=True)
data
```

```python
# Normalize positions relative to the first recorded position of each object
initial_positions = data.groupby('ObjectID').first().rename(columns={'X': 'InitX', 'Y': 'InitY', 'Z': 'InitZ'})
# initial_positions
data = data.join(initial_positions, on='ObjectID')
# data[['X', 'Y', 'Z']] = data[['X', 'Y', 'Z']] - data[['InitX', 'InitY', 'InitZ']]

# # Calculate velocity and acceleration
# data = data.sort_values(by=['ObjectID', 'Time'])
# data[['Vx', 'Vy', 'Vz']] = data.groupby('ObjectID')[['X', 'Y', 'Z']].diff() / data.groupby('ObjectID')['Time'].diff().dt.total_seconds()
# data[['Ax', 'Ay', 'Az']] = data.groupby('ObjectID')[['Vx', 'Vy', 'Vz']].diff() / data.groupby('ObjectID')['Time'].diff().dt.total_seconds()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial import KDTree
from scipy.cluster.vq import kmeans, vq
from scipy.interpolate import griddata
from scipy.stats import describe
from scipy.sparse import csr_matrix

# Load the Stanford Bunny dataset (replace with actual data loading code)
# For demonstration purposes, we'll create a simple synthetic dataset
bunny_points = np.random.rand(1000, 3)  # 1000 random 3D points

# Task 1: Clustering
centroids, _ = kmeans(bunny_points, 5)  # Cluster the points into 5 groups
cluster_labels, _ = vq(bunny_points, centroids)

# Task 2: Density Estimation
kdtree = KDTree(bunny_points)
distances, _ = kdtree.query(bunny_points, k=10)  # Compute distances to 10 nearest neighbors
local_densities = 1 / distances.mean(axis=1)  # Estimate local densities

# Task 3: Interpolation (for demonstration, we'll interpolate z-coordinate only)
grid_x, grid_y = np.meshgrid(np.linspace(min(bunny_points[:, 0]), max(bunny_points[:, 0]), 100),
                             np.linspace(min(bunny_points[:, 1]), max(bunny_points[:, 1]), 100))
grid_z = griddata((bunny_points[:, 0], bunny_points[:, 1]), bunny_points[:, 2], (grid_x, grid_y), method='linear')

# Task 4: Sparse Matrix Operations (for demonstration, we'll create a sparse matrix with random data)
sparse_matrix = csr_matrix(np.random.rand(100, 100))

# Task 5: Statistical Analysis
stats = describe(bunny_points, axis=0)

# Plotting (for visualization purposes)
fig = plt.figure(figsize=(12, 8))

# Plot original points colored by cluster label
ax1 = fig.add_subplot(231, projection='3d')
for i in range(5):
    ax1.scatter(bunny_points[cluster_labels == i, 0], bunny_points[cluster_labels == i, 1],
                bunny_points[cluster_labels == i, 2], label=f'Cluster {i+1}')
ax1.set_title('Clustered Points')
ax1.legend()

# Plot local densities
ax2 = fig.add_subplot(232, projection='3d')
ax2.scatter(bunny_points[:, 0], bunny_points[:, 1], bunny_points[:, 2], c=local_densities, cmap='viridis')
ax2.set_title('Local Densities')

# Plot interpolated surface
ax3 = fig.add_subplot(233, projection='3d')
ax3.plot_surface(grid_x, grid_y, grid_z, cmap='plasma')
ax3.set_title('Interpolated Surface')

# Plot sparse matrix
ax4 = fig.add_subplot(234)
ax4.spy(sparse_matrix, markersize=1)
ax4.set_title('Sparse Matrix')

# # Plot statistical summary
# ax5 = fig.add_subplot(235)
# ax5.bar(['Mean', 'Variance', 'Min', 'Max'], stats.mean, yerr=stats.variance, capsize=5)
# ax5.set_title('Statistical Summary')

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = 3*x + 2*y + 1 + np.random.normal(0, 1, 100)  # Plane with noise

# Matrix formulation
A = np.c_[x, y, np.ones(x.shape)]
C,_,_,_ = np.linalg.lstsq(A, z, rcond=None)  # Coefficients

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, color='blue')
# Create grid to plot the plane
xx, yy = np.meshgrid(np.linspace(min(x), max(x), 10), np.linspace(min(y), max(y), 10))
zz = C[0]*xx + C[1]*yy + C[2]
ax.plot_surface(xx, yy, zz, color='red', alpha=0.5)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

array = np.random.rand(1000, 5)
target = array[:, 0]

X_train, X_test, y_train, y_test = train_test_split(array, target, test_size=0.33, random_state=42)

pca = PCA(n_components=2)

pca.fit(X_train)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(121)
ax1.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train)


ax2 = fig.add_subplot(122)
ax2.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test)

```
